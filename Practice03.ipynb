Check if a list contains a palindrome of elements

```python
list = [1,2,3,2,1]
list_copy = list.copy()

if(list_copy == list) :
    print('List is palindrome')
else :
    print('List is not palindrom')
```

Given a list of numbers, remove all duplicates but keep the same order.

```python
list = [1, 2, 2, 3, 1, 4]
seen = set()
result = []
for i in list :
    if i not in seen :
        result.append(i)
        seen.add(i)

print(result)
```

Find biggest number and its index


```python
tup = (2,1,4,7,4,9,3)

big = max(tup)
i = tup.index(big)

print('Biggest Number is : ' , big)
print('Index is : ', i)
```
