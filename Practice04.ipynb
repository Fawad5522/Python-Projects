Enter marks of 3 subjects from the user and store them in dictionary.Use subject name as key and marks as value.

```python
sub = {}

phy = int(input('Enter Physics marks ' ))
sub.update({'physics' : phy})

math = int(input('Enter maths marks ' ))
sub.update({'maths' : math})

che = int(input('Enter chemistry marks ' ))
sub.update({'chemisty' : che})

print(sub)
```

Union and intersection set

```python
set1 = {1,2,3,4}
set2 = {2,3,4,5,6}

print(set1.union(set2)) # combines both set values and return new 

set1 = {1,2,3,4}
set2 = {2,3,4,5,6}

print(set1.intersection(set2)) #return commmon values
```
