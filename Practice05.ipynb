While Loops

Print numbers from 1 to 100.

```python
num = 1
while num <= 100 :
    print(num)
    num += 1
```

Print numbers from 100 to 1.

```python
num = 100
while num >= 1 :
    print(num)
    num -= 1
```

Print the multiplication table of a number n. 

```python
n = int(input('Enter any number to find its table ' ))
i = 1
while i <= 10 :
    print(n ,'*' ,i ,'=' , n * i )
    i += 1
    
```

print the list using loops

```python
list = [1,4,9,16,25,36,49,64,81,100]
idx = 0
while idx < len(list) :
    print(list[idx])
    idx += 1 
```

Search for a number x in this tuple using loops

```python
tup = (1,4,9,16,25,36,49,64,81,100)
x = 36
idx = 0

while idx < len(tup):
    if tup[idx] == x:
        print("Found number on index", idx)
    idx += 1
```

For Loops

Print the elements of the following list using a loop

```python
num = [1,3,4,54,3,23,4,6,9]

for value in num:
    print(value)
    
```

 Search for a number x in this tuple using loops

```python
tup = (1,4,9,16,25,36,49,64,81,100,49)
x = 49

idx = 0
for el in tup:
    if(el == x):
        print('Number found at index' ,idx)
        #break    (use break if you want to print 49 for a one time)
    idx += 1
 
```

Range()

Print the multiplication table of a number n. 

```python
n = int(input('Enter any number to find its table ' ))

for i in range(1 , 11):
    print(n ,'* ' ,i ,'=' ,n*i)
    i += 1
    
```

Print the elements of the following list using a loop

```python
num = [1,3,4,54,3,23,4,6,9]

for i in range(len(num)): #num is a list, and range() needs an integer
    print(i)
```

WAP ro find the sum of first natural num (using while)

```python
n = 7

sum = 0
i = 1
while i <= n:
    sum += i
    i += 1
    
print('Total Sum ' , sum)
```

WAP to find a factorial of natual numbers(Using for)

```python
n = 5

factorial = 1
for i in range(1,n+1):
    factorial *= i

print('Factorial of ',5 ,'is' ,factorial)
```
