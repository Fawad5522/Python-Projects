WAP to print the element of a list in a single line.(List is the parameter)

```python
heros = ['ironman' ,'thor', 'captain america']

def print_list(list):
    for item in list:
        print(item ,end = ' ')

print_list(heros)
```

WAP to find the factorial of n.(n is the parameter)

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i  
    print('Factorial of ',5 ,'is' ,fact)

factorial(5)
```

Recursive Function

Just to understand how its work.

```python
def show(n):
    if(n == 0): # base case is essential to know
        return    #otherwise program will crash
    print(n)
    show(n-1)  #call by itself 
    print('End')   

show(3)
```

Factorial through Recursive Function

```python
def fact(n):
    if(n == 0 or n == 1):
        return 1
    else:
        return n * fact(n-1)  # n! = (n-1)! * n  OR 5! = (5 -1)! * 5 = 4! * 5 = 120 

print(fact(5))
```

Calculate the sum of first n natural numbers through recursive function

```python
def calc_sum(n):
    if (n == 0):
        return 0
    return calc_sum(n-1) + n

sum = calc_sum(5)
print(sum)
```

Print all element in a list (use list and index as a parameter)

```python
def print_list(list ,idx):
    if(idx == len(list)):
        return

    print(list[idx])
    print_list(list, idx+1)

fruits = ['apple' ,'banana' ,'mango', 'avacado']
print_list(fruits,0)
```
