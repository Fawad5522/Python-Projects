Search if the word 'learning' exist in a file or not 

```python
def check_for_word():
    word = 'learning'
    with open('Practice.txt','r') as f:
        data = f.read()

    if(data.find(word)!= -1):
        print('Exist')

    else:
        print('Not Exist')

check_for_word()
```

WAF to find in which line of the file does the word 'learning' occur first.if not print -1

```python
def check_line():
    word = "learning"
    line = 1

    with open("Practice.txt", "r") as f:
        data = f.readline()
        while data:
            if word in data:
                return line
            line += 1
            data = f.readline()

    return -1

print(check_line())
```

From a file containing numbers separated by comma .print the count of even no

```python
f = open('Practice2.txt' ,'r')
f.close()
```

```python
count = 0
with open('Practice2.txt', 'r') as f:
    data = f.read()

    nums = data.split(',')
    for val in nums:
            if(int(val) % 2 == 0):
                count += 1
    
print(count)
```
